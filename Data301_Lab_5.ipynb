```
def cosine_similarity(a,b):
  """takes two vectors and returns the cosine similarity, 
  dot ab / len(a) times len(b)
  imports taken from numpy docs"""
  array_a = np.array(a)
  array_b = np.array(b)
  return np.dot(array_a, array_b) / (la.norm(array_a) * la.norm(array_b))

print("Pairs below should match")
print(cosine_similarity([1,2,-1],[2,1,1]))
print("0.5000000000000001")
print(cosine_similarity([1,2,-1],[1,2,1]))
print("0.6666666666666667")
print(cosine_similarity([1,2,1],[1,2,1]))
print("1.0000000000000002")
print(cosine_similarity([1, 1, 0, 1, 1, 0], [0, 0, 0, 0, 1, 1]))
print("0.35355339059327373")
```

```
#from sortedcontainers import SortedSet

def unique_reduction(destination_list):
  """Takes a list, reduces it to unique values and returns a set of each value and 1/len(unique_list)"""
  dictionary = {}
  destination_set = set(destination_list)
  split_value = 1/len(destination_set)
  for i in destination_set:
    dictionary[i] = split_value
  return dictionary

print("Below should match")
print(unique_reduction([1, 7, 7, 13, 13, 21, 25, 43, 52, 65, 66, 77, 77, 78, 79, 81, 94]))
print("{1: 0.07142857142857142, 65: 0.07142857142857142, 66: 0.07142857142857142, 7: 0.07142857142857142, 43: 0.07142857142857142, 13: 0.07142857142857142, 77: 0.07142857142857142, 78: 0.07142857142857142, 79: 0.07142857142857142, 81: 0.07142857142857142, 52: 0.07142857142857142, 21: 0.07142857142857142, 25: 0.07142857142857142, 94: 0.07142857142857142}")
```

```
#[10 points] Write a regular python function that takes a list of tuples and a vector and computes a single row’s sparse matrix multiplication. Here are some examples
def row_multiply(list_of_tuples, vector):
  result = 0
  for tuple_pair in list_of_tuples:
    index, value = tuple_pair
    result += value * vector[index]
  return result


print('Below pairs should match')
print(row_multiply([(0,1),(2,1)],[.2,.3,.5,0]))
print('0.7')
print(row_multiply([(1,1),(3,5)],[.2,.3,.5,0]))
print('0.3')
print(row_multiply([(0,.1),(1,.1),(2,.1),(3,.1)],[.2,.3,.5,0]))
print('0.1')
```
